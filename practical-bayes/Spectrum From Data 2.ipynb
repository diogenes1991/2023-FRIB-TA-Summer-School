# Spectrum from Data Challenge

The purpose of this challenge is for you to recover the energy spectrum of a nuclear decay by fitting a Breit-Wigner distribution to observed data:

## $\mathcal{F}(E, \alpha) =\frac{1}{2}\sum_{i=1}^n  \, \frac{A_i \, \Gamma _i}{(E-E_i)^2+\Gamma_i^2}$

This model is characterized by "n" peaks, each with center $E_i$, amplitude $A_i$, and width $\Gamma_i$. We group all these parameters into $\alpha=\{A, E, \Gamma \}$

The observed data and energies are defined below. Your taks are:

- Build your likelihood and prior as a function of "n" peaks (you will later vary n). For the likelihood be careful, we are modeling this as a Poisson process and you should use that accordingly (https://en.wikipedia.org/wiki/Poisson_distribution). 
- Do Metrolopis sampling to get posterior distributions for the parameters (centers, amplitudes, and widths) for three different models with 2, 3, and 4 peaks respectively. Plot the mean predictions with associated uncertainty bands.
- How many peaks do you think the spectrum has? What are your credible intervals for the associated parameters?
- Compare your results for n=3 with a calibration done with a misspecified Gaussian likelihood (write your likelihood as $e^{-\chi^2/2}$). 
- Estimate the influence of your posterior conclusions for n=3 from unaccounted background noise of the form: $B(E,\epsilon) = \epsilon \frac{1}{1+E^2}$. To do so, corrupt your data by adding the background term at every energy. Make various plots of the posterior distributions for different values of the background strenght $\epsilon$.




For an interesting paper related to this discussion that uses a cool technique (deblurring) check: https://arxiv.org/abs/2210.00157

```python
# Define the E range and number of data points
E_min = 0
E_max = 10

# Generate E values
EnGrid = np.linspace(E_min, E_max, 1000)
```

The following is the Breit-Wigner Spectrum function for a given set of parameters. 
It will create as many peaks as there are "centers" in the list.

```python
def Spectrum(xval,parameters):
    
    centers,amplitudes,widths=parameters
    
    value=0
    for i in range(len(centers)):
        value=value+ 1/2*amplitudes[i] * widths[i] / ((xval - centers[i]) ** 2 + widths[i] ** 2)
    return value
```

As we can see, these values of the parameters do not reproduce the data very well. Guess we will have to use Bayesian methods to find better values!

```python
class Metropolis:
    def __init__(self,**kwargs):
        pars = {}
        pars.update(**kwargs)
        
        self.history = []
        self.nsteps  = 0
        
        ''' Load the data to be put into the likelyhod '''
        self.data_x  = pars["data_x"]
        self.data_y  = pars["data_y"]
        
        ''' Load the characterization of the parameters '''
        self.model   = pars["model"]
        self.prior   = pars["prior"]
        self.lambdas = pars["lambdas"]
        
        ''' Jump function for choosing new parameters '''
        self.jump    = pars["jump"]
        
        ''' Misc '''
        self.data_l  = len(self.data_x)
        self.verbose = False
        
        
    def likelyhod(self,parameters):
        ''' 
            We assume the difference between observation and model
            follows a Poisson distribution with some lambda for each 
            observation
        '''
        rval = 1
        for i in range(self.data_l):
            counts = int(abs(self.data_y[i] - self.model(self.data_x[i],parameters)))
            factor = poisson.pmf(counts,lambdas[i])
            if self.verbose > 2:
                print("Counts",counts,"factor",factor)
            rval *= factor
        return rval
    
    def posterior(self,parameters):
        ''' Use Bayes' rule to compute non-normalized posteriors '''
        likelyhod = self.likelyhod(parameters)
        prior     = self.prior(parameters)
        if self.verbose > 1:
            print("likelyhood",likelyhod)
            print("prior",prior)
        return  likelyhod*prior 
    
    def run(self,**kwargs):
        
        pars = { "niter"   : 100,
                 "nrecord" : 10}
        pars.update(**kwargs)
        
        niter   = pars["niter"]
        nrecord = pars["nrecord"]
        
        ''' Initialize the Parameters '''
        self.state     = pars["start"]
        self.nsteps    = 0 
        self.numsucces = 0
        
        while self.nsteps < niter:
    
            ''' Propose change of state '''
            new_state = self.jump(self.state)
            
            ''' Generate a random number r '''
            rand = np.random.random()
            
            ''' Display stuff '''
            if self.verbose:
                print("Working on:",self.nsteps)
                if self.verbose > 1:
                    print("posterior(state)",self.posterior(self.state))
                    print("posterior(new state)",self.posterior(new_state))
                    print("r",rand)
            
            ''' Jump if the ratio of prosteriors is greater than r '''
            if self.posterior(new_state) > rand*self.posterior(self.state):
                self.state = new_state
                self.numsucces += 1
            
            ''' Record this state'''
            if self.nsteps > nrecord:
                self.history.append(self.state)
            self.nsteps += 1
        return
```

```python
''' Define our own Metropolis '''

numpars = len(Energies)
lambdas = [ 10 for i in range(numpars) ]

''' Centers '''
mu_centers = [ 0, 1.5, 4.5 ]
si_centers = [ 1**2, 1**2, 1**2 ]

''' Amplitudes '''
mu_amplitudes = [ 50, 25, 60 ] 
si_amplitudes = [ 10**2,  10**2, 10**2 ]

''' Widths '''
mu_widths = [ 0.35, 0.45, 2 ] 
si_widths = [ 1**2, 1**2, 1**2 ]

''' Sigma Matrix for the priors '''
mu = mu_centers + mu_amplitudes + mu_widths   
si = np.linalg.inv(np.diag(si_centers+si_amplitudes+si_widths))

''' Jump functions for all parameters '''
jumpsizes = [ 0.1,0.1,0.1, 2,2,2, 0.2,0.2,0.2 ]
jump = lambda state : np.array([ state[i] + np.random.normal(0,jumpsizes[i]) for i in range(9) ])

def Prior(parameters_vals,parameters_mu,parameters_sigma):
    ''' Gaussian priors '''
    arg   = np.array(parameters_vals) - np.array(parameters_mu)
    rval  = (2*np.pi)**(-len(parameters_vals)/2)
    rval *= np.sqrt(np.linalg.det(parameters_sigma))
    rval *= np.exp(-np.dot(arg,np.dot(parameters_sigma,arg))/2)
    return rval

def Spectrum(xval,parameters):
    
    ''' Model for us '''
    
    centers    = parameters[:3]
    amplitudes = parameters[3:6]
    widths     = parameters[6:]
    
    value=0
    for i in range(len(centers)):
        value= value + 1/2*amplitudes[i] * widths[i] / ((xval - centers[i]) ** 2 + widths[i] ** 2)
    return value

SpectrumMetropolis = Metropolis( data_x  = Energies,
                                 data_y  = counts,
                                 model   = Spectrum,
                                 prior   = lambda parameters : Prior(parameters,mu,si),
                                 lambdas = lambdas,
                                 jump    = jump ) 
```

```python
SpectrumMetropolis.verbose = 3
SpectrumMetropolis.run(start=mu,niter=10,nrecord=0)
```

```python
Mu_means = []
for i in range(len(SpectrumMetropolis.history[0])):
    Mu = [ h[i] for h in SpectrumMetropolis.history ] 
    Mu_means.append(Mu)
    plt.hist(Mu)
    plt.show()
    plt.clf()
```
